## triplets

```python
dataset = load_dataset('unicamp-dl/mmarco', 'portuguese')
```

```python
dataset["train"]
```

```python
dataset["train"][0]
```

```python
output_json = f'{mmarco_path}/triplets.jsonl'
output_tsv = f'{mmarco_path}/triplets.tsv'

with open(output_json,'w') as f_out:
    for line in tqdm(dataset["train"]):
        output_dict = {'query': line["query"], 'positive': line["positive"], 'negative': line["negative"]}
        f_out.write(json.dumps(output_dict, ensure_ascii=False) + '\n')

with open(output_tsv,'w') as f_out:
    for line in tqdm(dataset["train"]):
        f_out.write(f'{line["query"]}\t{line["positive"]}\t{line["negative"]}\n')
```

## qrel

```python
from collections import defaultdict
import json
import os

def build_json_qrel(qrel_file_path):
    """
    input file has format: 186154  0       1160    1
    """
    temp_d = defaultdict(dict)
    with open(qrel_file_path) as reader:
        for line in reader:
            q_id, _, d_id, rel = line.split("\t")
            temp_d[q_id][d_id] = int(rel)
    print(f"built qrel file, contains {len(temp_d)} queries...")
    json.dump(dict(temp_d), open(os.path.join(os.path.dirname(qrel_file_path), "qrel_dev.json"), "w"))
```

```python
build_json_qrel("data/raw_data/qrels.dev.tsv")
```

## val_collection based on qrels

```python
import pandas as pd
import json

with open('data/raw_data/qrel_dev.json') as f:
    qrels = json.load(f)

# Collect doc_ids from qrels
doc_ids_qrels = [int(item) for sublist in qrels.values() for item in sublist]
```

```python
len(doc_ids_qrels)
```

```python
doc_ids_qrels[:10]
```

```python
collection = pd.read_csv('data/raw_data/portuguese_collection.tsv', sep='\t', header=None, names=['doc_id', 'text'])
```

```python
collection.tail()
```

```python
validation_collection = collection[collection['doc_id'].isin(doc_ids_qrels)]
```

```python
validation_collection.head()
```

```python
validation_collection.to_csv('data/raw_data/validation_collection.tsv', sep='\t', index=False, header=False)
```

```python
# Exclude rows that are already in validation_collection.tsv
remaining_collection = collection[~collection['doc_id'].isin(validation_collection['doc_id'])]
```

```python
# Sample additional 100,000 documents from the remaining_collection
additional_samples = remaining_collection.sample(n=100000, random_state=123)  # random_state for reproducibility
```

```python
# Append additional_samples to validation_collection
validation_collection = pd.concat([validation_collection, additional_samples])
```

```python
len(validation_collection)
```

```python
# Save to validation_collection.tsv
validation_collection.to_csv('validation_collection.tsv', sep='\t', index=False, header=False)
```

## val_collection based on val_queries

```python
import pandas as pd
import json
```

```python
# Step 1: Load val_queries.tsv and get validation query IDs
val_queries = pd.read_csv('data/raw_data/queries.dev.small.tsv', sep='\t', header=None, names=['query_id', 'text'])
val_queries.head()
```

```python
val_query_ids = val_queries['query_id'].tolist()
val_query_ids[:10]
```

```python

# Step 2: Load qrels.json and get doc_ids associated with validation query IDs
with open('data/raw_data/qrel_dev.json') as f:
    qrels = json.load(f)

# Collect doc_ids associated with validation query IDs
val_doc_ids = [int(item) for key, sublist in qrels.items() if int(key) in val_query_ids for item in sublist]
val_doc_ids[:10]
```

```python
len(val_doc_ids)
```

```python
# Step 3: Load collection.tsv
collection = pd.read_csv('data/raw_data/portuguese_collection.tsv', sep='\t', header=None, names=['doc_id', 'text'])

# Step 4: Filter collection based on validation doc_ids and save to validation_collection.tsv
validation_collection = collection[collection['doc_id'].isin(val_doc_ids)]
validation_collection.head()
```

```python
# Exclude rows that are already in validation_collection.tsv
remaining_collection = collection[~collection['doc_id'].isin(validation_collection['doc_id'])]
```

```python
# Sample additional 10000 documents from the remaining_collection
additional_samples = remaining_collection.sample(n=10000, random_state=123)  # random_state for reproducibility
```

```python
# Append additional_samples to validation_collection
validation_collection = pd.concat([validation_collection, additional_samples])
len(validation_collection)
```

```python
validation_collection.to_csv('validation_collection_on_val_queries_small.tsv', sep='\t', index=False)
```
